```python
import pandas as pd

data = {
        'Valore': [10, 5, 8, 12, 3],
        'Quantità': [20, 15, 10, 25, 5],
        'Testo': ['abc', 'def', 'ghi', 'jkl', 'mno'],
        'Categoria': ['A', 'B', 'A', 'C', 'B'],
        'ID': [1, 2, 3, 4, 1],
        'Descrizione': ['desc1 importante', 'desc2', 'desc3', 'desc4 importante', 'desc5']
        }

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
def doppio(x):
    return x*2
```

```python
df['Valore'].apply(doppio)
```

```python
df['Valore'].apply(lambda x: x *2)
```

```python
## Creare una nuova colonna in un DataFrame chiamata "Doppio_Valore" che contenga il doppio del valore della colonna esistente "Valore".
df['Doppio_Valore'] = df['Valore'].apply(lambda x: x * 2)
df
```

```python
#Filtrare le righe di un DataFrame in base a una condizione, ad esempio, ottenere solo le righe con valori superiori a 3 nella colonna "Valore".
df[df['Valore'].apply(lambda x: x > 3)]
```

```python
#Applicare una lambda function alla colonna "Testo" per convertire tutti i valori in maiuscolo.
df['Testo'] = df['Testo'].apply(lambda x: x.upper())
df
```

```python
#Contare quante volte compare ciascun valore nella colonna "Categoria".

conteggio_categorie = df.groupby(lambda x: df.loc[x, 'Categoria']).size().reset_index()
conteggio_categorie.columns = ['Categoria', 'Conteggio']
conteggio_categorie
```

```python
#Rinominare tutte le colonne di un DataFrame aggiungendo il prefisso "Nuovo_".
df.rename(columns=lambda x: f'Nuovo_{x}')
```

```python
#Applicare una funzione a tutte le colonne:
df.apply(lambda col: col.sum())
```

```python
#Ordinare il DataFrame in base alla colonna "Valore" in ordine decrescente.

df.sort_values(by='Valore', ascending=False)
```

```python
#Creare una nuova colonna "Totale" che rappresenti la somma di "Valore" e "Quantità".

df['Totale'] = df.apply(lambda row: row['Valore'] + row['Quantità'], axis=1)
df
```

```python
#Rimuovere le righe duplicate basandosi sulla colonna "ID".
df.drop_duplicates(subset='ID')
```

```python
#Filtrare le righe del DataFrame in cui la colonna "Descrizione" contiene la parola "importante".
df[df['Descrizione'].apply(lambda x: 'importante' in x.lower())]
```
