```python
import pandas as pd

# Creare due DataFrame di esempio
df1 = pd.DataFrame({
                    'ID': [1, 2, 3, 4],
                    'Nome': ['Alice', 'Bob', 'Charlie', 'David'],
                    'Età': [25, 30, 35, 40]
                    })

df2 = pd.DataFrame({
                    'ID': [2, 3, 4, 5,4],
                    'Stipendio': [50000, 60000, 70000, 80000,70000],
                    'Posizione': ['Analista', 'Manager', 'Direttore', 'CEO','Direttore']
                    })
```

```python
df1
```

```python
df2
```

```python
# 1.Esegui una join (merge) tra i due DataFrame utilizzando l'ID come chiave.
merged_df = pd.merge(df1, df2, on='ID')
merged_df
```

```python
# 2. Esegui una join esterna (outer) tra i due DataFrame.
outer_merged_df = pd.merge(df1, df2, on='ID', how='outer')
outer_merged_df
```

```python
# 3. Esegui una join sinistra (left merge) tra i due DataFrame.
left_merged_df = pd.merge(df1, df2, on='ID', how='left')
left_merged_df
```

```python
# 4. Esegui una join destra (right merge) tra i due DataFrame.
right_merged_df = pd.merge(df1, df2, on='ID', how='right')
right_merged_df
```

```python
# 5. Esegui una join su più colonne (utilizzando una lista di chiavi).
multi_key_merged_df = pd.merge(df1, df2, on=['ID'], how='inner')
multi_key_merged_df
```

```python
# 6. Esegui una join su colonne diverse nei due DataFrame.
custom_merge_df = pd.merge(df1, df2, left_on='ID', right_on='ID', how='inner')
custom_merge_df
```

```python
# 7. Gestisci valori duplicati durante la join.
df2_clean = df2.drop_duplicates()
merged_duplicates_df = pd.merge(df1, df2_clean, on='ID', how='inner')
merged_duplicates_df
```

```python
import pandas as pd

df3 = pd.DataFrame({
                    'ID': [1, 2, 3, 4],
                    'Progetto': ['A', 'B', 'C', 'D'],
                    'Durata': [10, 15, 20, 25]
                    })

df4 = pd.DataFrame({
                    'ID': [1, 2, 3, 5],
                    'Progetto': ['A', 'B', 'C', 'E'],
                    'Costo': [1000, 1500, 2000, 2500]
                    })
```

```python
df3
```

```python
df4
```

```python
df4[df4['Costo'] > 1500]
```

```python
# 8. Esegui una join con una condizione di filtro basata su una colonna. df3,df4 costo>1500
filtered_merge_df = pd.merge(df3, df4[df4['Costo'] > 1500], on='ID', how='inner')
filtered_merge_df
```

```python
# 9. Esegui una join considerando una colonna come indice in uno dei DataFrame df3,df4
df4.set_index('ID', inplace=True)
index_merge_df = pd.merge(df3, df4, left_on=['ID'], right_index=True, how='inner')
index_merge_df
```

```python
import pandas as pd

# Esercizio 1: Concatenare due DataFrame x righe
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

result1 = pd.concat([df1, df2]).reset_index(drop=True)
result1
```

```python
# Esercizio 2: Concatenare due DataFrame lungo le colonne
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

result2 = pd.concat([df1, df3], axis=1)
result2
```

```python
# Esercizio 3: Concatenare più DataFrame lungo le righe
df4 = pd.DataFrame({'A': [13, 14], 'B': [15, 16]})

result3 = pd.concat([df1, df2, df4]).reset_index(drop=True)
result3
```

```python
# Esercizio 4: Concatenare più DataFrame lungo le colonne
df5 = pd.DataFrame({'C': [17, 18], 'D': [19, 20]})

result4 = pd.concat([df1, df3, df5], axis=1)
result4
```

```python
dfx = pd.DataFrame({'A': [17, 18], 'B': [19, 20]})

resultx = pd.concat([df1, df2, df5], axis=0).reset_index(drop=True)
resultx
```

```python
# Esercizio 5: Concatenare DataFrame con indici diversi
df6 = pd.DataFrame({'A': [21, 22], 'B': [23, 24]}, index=[0, 1])
df7 = pd.DataFrame({'A': [25, 26], 'B': [27, 28]}, index=[2, 3])

result5 = pd.concat([df6, df7])
result5
```

```python
# Esercizio 6: Ignorare gli indici esistenti durante la concatenazione di df6 e df7
result6 = pd.concat([df6, df7], ignore_index=True)
result6
```

```python
# Esercizio 7: Specificare le chiavi per identificare le origini dei dati
result7 = pd.concat([df1, df2], keys=['df1', 'df2'])
result7
```

```python
# Esercizio 8: Concatenare DataFrame df1 con df8 con colonne diverse
df8 = pd.DataFrame({'A': [29, 30], 'E': [31, 32]})
result8 = pd.concat([df1, df8], axis=1)
result8
```

```python
# Esercizio 9: Concatenare DataFrame df9 con df6 mantenendo le colonne
df9 = pd.DataFrame({'A': [33, 34], 'B': [35, 36]}, index=[1, 2])
result9 = pd.concat([df6, df9], axis=1, join='outer', keys=['df6', 'df9'])
result9
```
