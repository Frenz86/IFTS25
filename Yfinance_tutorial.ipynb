### Intraday plot

```python
hist=tsla.history(start='2024-04-06',
                  end='2024-05-08',
                  #interval='1m',
                  actions=False)
hist
```

```python
mpf.plot(hist, type='candle', volume=True, style='yahoo', mav=(20),
        title='TSLA 2023-05-12', tight_layout=True)
```

### Fetching multiple tickers

```python
data = yf.download("AMZN AAPL TSLA",
                   period="ytd",
                   group_by='ticker',
                   actions=False)
amzn=data["AMZN"]
aapl=data["AAPL"]
```
