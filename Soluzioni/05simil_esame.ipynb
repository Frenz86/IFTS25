```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wine.csv')
```

```python
df_clean = df.copy()
```

```python
# Rimuovi Ask For Price e reset index
df_clean = df_clean[df_clean['Price'] != 'Ask For Price'].reset_index(drop=True)
```

```python
# rimuovi virgola in prices
df_clean['Price'] = df_clean['Price'].str.replace(',', '').astype(int)
```

```python
# prezzi hanno scale non reali
df_clean['Price'] = df_clean['Price']/100
```

```python
df_clean['bottles_stored'] = df_clean['bottles_stored'].replace('', np.nan)
df_clean['bottles_stored'] = df_clean['bottles_stored'].str.replace(',', '')
df_clean['bottles_stored'] = df_clean['bottles_stored'].str.replace('bottles', 'btl')
df_clean['bottles_stored'] = df_clean['bottles_stored'].str.extract('(\d+)').astype(float)
```

```python
# year come intero
df_clean['year'] = pd.to_numeric(df_clean['year'], errors='coerce').astype(int)
```

```python
# gli anni più frequenti

mask = df_clean['year'].value_counts().nlargest(20).index
df_clean = df_clean[df_clean['year'].isin(mask)].reset_index(drop=True)
```

```python
# Estrazione del tipo vino dal nome
df_clean['wine_type'] = df_clean['name'].str.split().str[0]
```

```python
# Calculate age
df_clean['age_years'] = 2024 - df_clean['year']
```

```python
# Fill cellar_type con la moda
df_clean['cellar_type'] = df_clean['cellar_type'].fillna(df_clean['cellar_type'].mode()[0])
```

```python
#drop columns and nan
df_clean = df_clean.drop(columns='Unnamed: 0')
df_clean = df_clean.dropna()
```
