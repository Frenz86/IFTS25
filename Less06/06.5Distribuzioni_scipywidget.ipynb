```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, triang, norm
```

```python
# Generazione dei campioni casuali
num_samples = 1000
```

```python
# Distribuzione costante
min_const = 0
max_const = 10
const_dist = uniform(loc=min_const, scale=max_const-min_const)


const_samples = const_dist.rvs(size=num_samples)

# Distribuzione costante
plt.hist(const_samples, bins=20, density=True, alpha=0.6, color='g')
x_const = np.linspace(min_const, max_const, 100)
plt.plot(x_const, const_dist.pdf(x_const), 'k-', lw=2, label='PDF')
plt.title('Distribuzione Costante')
plt.legend(loc=1);
```

```python
from scipy.stats import triang
# Distribuzione triangolare
c = 0.5  # Moda percent
min_tri = 2
max_tri = 12
tri_dist = triang(c, loc=min_tri, scale=max_tri-min_tri)

tri_samples = tri_dist.rvs(size=num_samples)

# Distribuzione triangolare
plt.hist(tri_samples, bins=20, density=True, alpha=0.6, color='b')
x_tri = np.linspace(min_tri, max_tri, 100)
plt.plot(x_tri, tri_dist.pdf(x_tri), 'k-', lw=2, label='PDF')
plt.title('Distribuzione Triangolare')
plt.legend(loc=1);
```

```python
# Distribuzione gaussiana

mu = 0  # Media
sigma = 1  # Deviazione standard
gauss_dist = norm(loc=mu, scale=sigma)

gauss_samples = gauss_dist.rvs(size=num_samples)

# Distribuzione gaussiana
plt.hist(gauss_samples, bins=20, density=True, alpha=0.6, color='r')
x_gauss = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x_gauss, gauss_dist.pdf(x_gauss), 'k-', lw=2, label='PDF')
plt.title('Distribuzione Gaussiana')
plt.legend(loc=1);
```

## Widget con le Distribuzioni

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
import ipywidgets as widgets

# Creazione dei widget interattivi per modificare i parametri
min_const_widget = widgets.FloatSlider(value=0, min=-10, max=10, step=0.1, description='Min Const:')
max_const_widget = widgets.FloatSlider(value=10, min=-10, max=20, step=0.1, description='Max Const:')

# Funzione di aggiornamento del plot
def update_plot(min_const, max_const):
    # Creazione della distribuzione costante
    const_dist = uniform(loc=min_const, scale=max_const-min_const)

    # Generazione dei campioni casuali
    num_samples = 1000
    const_samples = const_dist.rvs(size=num_samples)

    # Plot della distribuzione costante
    plt.figure(figsize=(8, 6))
    plt.hist(const_samples, bins=20, density=True, alpha=0.6, color='g')
    x_const = np.linspace(min_const, max_const, 100)
    plt.plot(x_const, const_dist.pdf(x_const), 'k-', lw=2, label='PDF')
    plt.title('Distribuzione Costante')
    plt.xlabel('Valore')
    plt.ylabel('Densit di probabilit')
    plt.legend()
    plt.grid(True)
    plt.show()

# Connessione dei widget all'evento di aggiornamento
widgets.interactive(update_plot, min_const=min_const_widget, max_const=max_const_widget)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang
import ipywidgets as widgets

# Creazione dei widget interattivi per modificare i parametri
c_widget = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.1, description='Moda Percentuale:')
min_tri_widget = widgets.FloatSlider(value=0, min=-10, max=10, step=0.1, description='Min Tri:')
max_tri_widget = widgets.FloatSlider(value=10, min=-10, max=20, step=0.1, description='Max Tri:')

# Funzione di aggiornamento del plot
def update_plot(c, min_tri, max_tri):
    # Creazione della distribuzione triangolare
    tri_dist = triang(c, loc=min_tri, scale=max_tri-min_tri)

    # Generazione dei campioni casuali
    num_samples = 1000
    tri_samples = tri_dist.rvs(size=num_samples)

    # Plot della distribuzione triangolare
    plt.figure(figsize=(8, 6))
    plt.hist(tri_samples, bins=20, density=True, alpha=0.6, color='b')
    x_tri = np.linspace(min_tri, max_tri, 100)
    plt.plot(x_tri, tri_dist.pdf(x_tri), 'k-', lw=2, label='PDF')
    plt.title('Distribuzione Triangolare')
    plt.xlabel('Valore')
    plt.ylabel('Densit di probabilit')
    plt.legend()
    plt.axis([0, 10, 0, 0.3])
    plt.grid(True)
    plt.show()

# Connessione dei widget all'evento di aggiornamento
widgets.interactive(update_plot, c=c_widget, min_tri=min_tri_widget, max_tri=max_tri_widget)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets

# Creazione dei widget interattivi per modificare i parametri
mu_widget = widgets.FloatSlider(value=0, min=-10, max=10, step=0.1, description='Media:')
sigma_widget = widgets.FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Dev. Std.:')

# Funzione di aggiornamento del plot
def update_plot(mu, sigma):
    # Creazione della distribuzione gaussiana
    gauss_dist = norm(loc=mu, scale=sigma)

    # Generazione dei campioni casuali
    num_samples = 1000
    gauss_samples = gauss_dist.rvs(size=num_samples)

    # Plot della distribuzione gaussiana
    plt.figure(figsize=(8, 6))
    plt.hist(gauss_samples, bins=20, density=True, alpha=0.6, color='r')
    x_gauss = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    plt.plot(x_gauss, gauss_dist.pdf(x_gauss), 'k-', lw=2, label='PDF')
    plt.title('Distribuzione Gaussiana')
    plt.xlabel('Valore')
    plt.ylabel('Densit di probabilit')
    plt.legend()
    plt.axis([-10, 10, 0, 0.5])
    plt.grid(True)
    plt.show()

# Connessione dei widget all'evento di aggiornamento
widgets.interactive(update_plot, mu=mu_widget, sigma=sigma_widget)
```
