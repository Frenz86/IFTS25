## *Args per le funzioni
L'argomento *args in una funzione Python per sommare tutti i numeri passati come argomenti.

```python
somma = 0
numeri = [1, 2, 3, 4, 5]

for numero in numeri:
    somma += numero
print("La somma è:", somma)
```

```python
def somma_numeri(*args):
    somma = 0
    for numero in args:
        somma += numero
    return somma
```

```python
print(somma_numeri(1, 2, 3, 4, 5))  # Stampa 15
print(somma_numeri(10, 20, 30))      # Stampa 60
print(somma_numeri(2, 2, 2, 2, 2))   # Stampa 10
```

```python
def concatena_stringhe(*args):
    stringa_concat = ""
    for stringa in args:
        stringa_concat += stringa
    return stringa_concat
```

```python
# Esempi di utilizzo della funzione
print(concatena_stringhe("Ciao, ", "come ", "stai?"))  # Stampa "Ciao, come stai?"
print(concatena_stringhe("Queste ", "sono ", "stringhe ", "concatenate."))  # Stampa "Queste sono stringhe concatenate."
```

# Classi
Le classi sono un meccanismo fondamentale per la creazione di oggetti e l'organizzazione del codice. Quando si definisce una classe, è possibile includere il parametro "self" come primo parametro all'interno dei metodi di classe.

"self" è una convenzione in Python per fare riferimento all'istanza corrente di una classe. Quando si definisce un metodo all'interno di una classe e si passa "self" come primo parametro, si sta indicando che il metodo opera sull'istanza corrente di quella classe.

Ad esempio, consideriamo la seguente classe chiamata "Persona":

```python
class XXXXX:
    def __init__(self,a,b): # ATTRIBUTI
        self.attr1 = a
        self.attr2 = b
```

```python
class Persona:
    def __init__(self,a,b): # ATTRIBUTI
        self.nome = a
        self.eta = b

    def saluta(self): # METHOD
        print("Ciao, mi chiamo", self.nome)

    def double(self,x): # METHOD
        ''' moltiplica l'input x2 volte'''
        return x*2

    def triple(self,x): # METHOD
        ''' moltiplica l'input x3 volte'''
        return x*3

    def quadratoanni(self): # METHOD
        ''' RESTITUISCE IL QUADRATO DEGLI ANNI'''
        return self.eta**2
```

```python
class Prodotto:
    iva = 0.22  # attributo di classe

    def __init__(self, nome, prezzo_netto):
        self.nome = nome
        self.prezzo_netto = prezzo_netto

    def prezzo_con_iva(self):
        return self.prezzo_netto * (1 + self.iva)

    def __str__(self):
        return f"{self.nome}: {self.prezzo_con_iva():.2f}€"

class ProdottoAlimentare(Prodotto):
    iva = 0.10  # IVA ridotta per alimenti

    def __init__(self, nome, prezzo_netto, data_scadenza):
        super().__init__(nome, prezzo_netto)
        self.data_scadenza = data_scadenza

    def __str__(self):
        return f"{self.nome}: {self.prezzo_con_iva():.2f}€ (Scade: {self.data_scadenza})"
```

```python
# Creare un negozio con vari prodotti
prodotti = [
            Prodotto("Laptop", 800),
            ProdottoAlimentare("Pasta", 1.5, "2023-12-31"),
            Prodotto("Smartphone", 500),
            ProdottoAlimentare("Latte", 1.2, "2023-06-15"),
            ]

print("Catalogo prodotti:")
for prodotto in prodotti:
    print(prodotto)
```
